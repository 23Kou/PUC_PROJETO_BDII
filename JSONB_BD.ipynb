```python
def executar_consulta(sql):
    cur.execute(sql)
    resultados = cur.fetchall()
    for linha in resultados:
        print(linha)
    print("-" * 60)
```

Consulta 1 — -> pega um valor do JSON como JSON

```python
executar_consulta("""
SELECT nome, detalhes -> 'especificacoes'
FROM produtos;
""")
```

Consulta 2 — ->> pega um valor do JSON como texto

```python
executar_consulta("""
SELECT nome, detalhes ->> 'categoria'
FROM produtos;
""")
```

Consulta 3 — #> pega um caminho dentro do JSON

```python
executar_consulta("""
SELECT nome, detalhes #> '{especificacoes}'
FROM produtos;
""")
```

Consulta 4 — #>> pega o valor final em texto

```python
executar_consulta("""
SELECT nome, detalhes #>> '{especificacoes,preco}'
FROM produtos;
""")
```

Consulta 5 — @> verifica contenção

```python
executar_consulta("""
SELECT nome
FROM produtos
WHERE detalhes @> '{"categoria":"audio"}'::jsonb;
""")
```

Consulta 6 — <@ verifica contenção reversa

```python
executar_consulta("""
SELECT nome
FROM produtos
WHERE '{"categoria":"audio"}'::jsonb <@ detalhes;
""")
```

Consulta 7 — ? verifica existência de chave

```python
executar_consulta("""
SELECT nome
FROM produtos
WHERE detalhes ? 'tags';
""")
```

Consulta 8 — ?| verifica se existe pelo menos uma chave da lista

```python
executar_consulta("""
SELECT nome
FROM produtos
WHERE detalhes ?| array['tags', 'avaliacoes'];
""")
```

Consulta 9 — ?& verifica se todas as chaves existem

```python
executar_consulta("""
SELECT nome
FROM produtos
WHERE detalhes ?& array['tags', 'avaliacoes'];
""")
```

Consulta 10 — jsonb_path_exists e jsonb_path_query

```python
executar_consulta("""
SELECT nome
FROM produtos
WHERE jsonb_path_exists(detalhes, '$.avaliacoes[*] ? (@.nota >= 5)');
""")
```

```python
executar_consulta("""
SELECT nome, jsonb_path_query(detalhes, '$.avaliacoes[*].nota')
FROM produtos;
""")
```
